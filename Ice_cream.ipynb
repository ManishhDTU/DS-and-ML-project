# Second Method for the analysis of Ice cream data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Ice%20Cream.csv")
```

```python
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.columns
```

```python
data.dtypes
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.boxplot(column =["Temperature"])
```

```python
data.boxplot(["Revenue"])
```

```python
data.boxplot(column=["Temperature", "Revenue"])
```

```python
# Dropping Serial No columnsas it is not required for prediction
data1 = data.copy()
data1
```

```python
# Calculating Quartiles of each feature
Q1 = data1.quantile(0.25)
Q3 = data1.quantile(0.75)
IQR = Q3 -Q1
print(IQR)
```

```python
# Identifying Outliers
df_out1 = data1[((data1 < (Q1 - 1.5 * IQR)) | (data1 > (Q3 + 1.5 * IQR))).any(axis=1)]
df_out1
```

```python
# removing outliers
df_out = data1.loc[~((data1 < (Q1 - 1.5 * IQR)) | (data1 > (Q3 + 1.5 * IQR))).any(axis=1)]
data = df_out.copy()
print(data.shape)
```

```python
data["Temperature"].plot.hist()
plt.xlabel("Temp.", fontsize = 14)
```

```python
data["Revenue"].plot.hist()
plt.xlabel("Revenue", fontsize = 14)
```

```python
data.plot.scatter("Temperature", "Revenue")
```

```python
data["Temperature"].corr(data["Revenue"])
```

```python
# separting independent and dependent variable
x = data[["Revenue"]]
y = data["Temperature"]
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```
