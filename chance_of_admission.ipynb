# Chance of Admission Prediction


## Predict the chances of admission of a student to a Graduate program based on:

#### GRE Scores (290 to 340)
#### TOEFL Scores (92 to 120)
#### University Rating (1 to 5)
#### Statement of Purpose (1 to 5)
#### Letter of Recommendation Strength (1 to 5)
#### Undergraduate CGPA (6.8 to 9.92)
#### Research Experience (0 or 1)
#### Chance of Admit (0.34 to 0.97)

```python
# step 1: import library
import pandas as pd
```

```python
# step 2: import data
admission = pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Admission%20Chance.csv")
```

```python
admission
```

```python
admission.head(7)
```

```python
admission.info()
```

```python
admission.describe()
```

```python
# step 3: define target(y) and features(x)
```

```python
admission.columns
```

```python
y = admission["Chance of Admit "]
y.shape
```

```python
x = admission.drop(["Serial No","Chance of Admit "],axis=1)
x.shape
```

```python
"""
x = admission[['GRE Score', 'TOEFL Score', 'University Rating', ' SOP',
       'LOR ', 'CGPA', 'Research']]
x
"""
# both above and this statement are same
```

```python
# Step 4: train_test_split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=.75, random_state = 2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5:
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model
```

```python
# Step 6
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Step 7: predict model
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# Step 8: model accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```
